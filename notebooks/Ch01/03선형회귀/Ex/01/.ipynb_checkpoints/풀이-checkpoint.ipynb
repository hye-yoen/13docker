```python
# ---------------------------------
# 데이터 전처리
# ---------------------------------
```

```python
target = train.pop('SalePrice')
```

```python
train.shape , test.shape
```

```python
# 중요도 있는 컬럼 선별
features = [
    "OverallQual",   # 전반적 자재/마감 품질
    "GrLivArea",     # 지상 생활 면적
    "GarageCars",    # 차고 수용 차량 수
    "GarageArea",    # 차고 면적
    "TotalBsmtSF",   # 지하 전체 면적
    "1stFlrSF",      # 1층 면적
    "FullBath",      # 전체 욕실 개수
    "YearBuilt",     # 건축 연도
    "YearRemodAdd",  # 리모델링 연도
    "KitchenQual",   # 주방 품질
    "Fireplaces",    # 벽난로 개수
    "Neighborhood"   # 위치
]
type(train[features])
train = train[features]
test = test[features]
```

```python
# train.info()
```

```python
test.info()
```

```python
# 범주형 -> 최빈값 # 수치형 : 최소값 (차 못넣는 사람 있을 가능성)
```

```python
#수치형 결측치 채우기
test.describe()
test['GarageCars'] = test['GarageCars'].fillna(test['GarageCars'].min())
test['GarageArea'] = test['GarageArea'].fillna(test['GarageArea'].min())
test['TotalBsmtSF'] = test['TotalBsmtSF'].fillna(test['TotalBsmtSF'].min())
```

```python
#범주형 결측치 채우기 (최빈값/ 사분위수 범위내 중 최소값/최대값)
# test['KitchenQual'].mode()[0]
# Ex = 우수 Gd = 좋음 TA 보통 Fa = 나쁨
test['KitchenQual'] = test['KitchenQual'].fillna(test['KitchenQual'].mode()[0])
test.isnull().sum()
```

```python
#이상치 생략
```

```python
#인코딩(LIGHTGBM 사용시 범주형(Object) -> 카테고리(category) 자료형)
train['KitchenQual'] = train['KitchenQual'].astype('category')
train['Neighborhood'] = train['Neighborhood'].astype('category')
test['KitchenQual'] = test['KitchenQual'].astype('category')
test['Neighborhood'] = test['Neighborhood'].astype('category')
test.info()
```

```python
# ---------------------------------
# 검증 데이터 분할 train - val (8,2)
# ---------------------------------
```

```python
from sklearn.model_selection import train_test_split
X_train , x_val, y_train, y_val = train_test_split(
    train,
    target,
    test_size=0.2,
    random_state=0
)
X_train.shape , x_val.shape, y_train.shape, y_val.shape
```

```python
# ---------------------------------
# 학습 및 평가(MSE, MAE , R2..)
# ---------------------------------
```

```python
#학습
import lightgbm as lgb
model = lgb.LGBMRegressor(random_state=0 , verbose=-1)
model.fit(X_train,y_train) #학섭
y_pred = model.predict(x_val) #예측
# y_pred
```

```python
#평가
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error


lr_mse = mean_squared_error(y_val , y_pred)
lr_mae = mean_absolute_error(y_val , y_pred)
lr_rmse = root_mean_squared_error(y_val, y_pred)
lr_r2 = r2_score(y_val,y_pred)

print("MSE : ",lr_mse ) #오차를 제곱해 쳥균을 낸 값 - 값이 클수록 예측이 실제와 다름
print("MAE : ",lr_mae ) #오차를 젤대값으로 계산- 값이 클수록 예측이 실제와 다름
print("RMSE : ",lr_rmse)
print("R2 : ",lr_r2)
```

```python
#결과 평가
# MSE :  1138255660.1912904 -> 다른 모델과 비교시 사용(튜닝이전/이후 비교) 낮은 값 코드 사용
# MAE :  19091.970142754628 -> 예측값과 실제 집값의 차이가 평균적으로 19000달러
# RMSE :  33738.044700179205 -> 얘측값과 실제 집값의 차이가 33000 - 약 20%정도 정도 오차
# R2 :  0.8351751535112408   -> 모델이 집값에 대한 예측을 83.5%정도 설명이 가능(우수한 경로)
```
